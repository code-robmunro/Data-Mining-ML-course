```python
import pandas as pd
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames, engine='python')

# create one merged DataFrame
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
```

```python
# Aggregate movie ratings for genres
import numpy as np
movie_stats = lens.groupby('genres').agg({'rating': [numpy.size, numpy.mean]})
movie_stats.head()
movie_stats.sort_values([('rating', 'mean')], ascending=False).head(15)
```

```python
# Top 5 ranked genres for females
mean_ratings = lens.pivot_table('rating', index='genres', columns='gender', aggfunc='mean')
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)

top_female_ratings[:5]
```

```python
# Top 5 ranked genres for males
top_male_ratings = mean_ratings.sort_values(by='M', ascending=False)

top_male_ratings[:5]
```

```python
# Movie rating by age for A Space Odyssey (1968)
most_50 = lens.groupby('movie_id').size().sort_values(ascending=False)[:50]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
lens['age_group'] = pd.cut(lens.age, range(0, 81, 10), right=False, labels=labels)
lens[['age', 'age_group']].drop_duplicates()
lens.groupby('age_group').agg({'rating': [np.size, np.mean]})
lens.set_index('movie_id', inplace=True)
by_age = lens.loc[most_50.index].groupby(['title', 'age_group'])
by_age.rating.mean().head(6)
```
