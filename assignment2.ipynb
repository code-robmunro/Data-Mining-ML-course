#### 1 Iris Dataset


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fileURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(fileURL, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Name"], header=None)
```

#### 1.1 Summary Statistics

```python
# Print the first 5 elements
df.head(5)
```

##### There are four features and one class.
##### The four features are numeric (interval) and measure physical features of two parts of iris flowers, the sepal and petals. Each part has a length and width.
##### The last column is categorical (nominal) and describes the class of iris flower associated with this instance.

```python
# Sepal ########

print('Sepal Length:\n')
print(df['Sepal Length'].describe())

# Calculate the variance (square of std) and print it
print('variance: ' + str(df['Sepal Length'].var()))

# Calculate the range (max-min) and print it
print('range: ' + str(df['Sepal Length'].ptp()))


print('\n\nSepal Width:\n')
print(df['Sepal Width'].describe())

# Calculate the variance (square of std) and print it
print('variance: ' + str(df['Sepal Width'].var()))

# Calculate the range (max-min) and print it
print('range: ' + str(df['Sepal Width'].ptp()))


# Petal ########
print('\n\nPetal Length:\n')
print(df['Petal Length'].describe())

# Calculate the variance (square of std) and print it
print('variance: ' + str(df['Petal Length'].var()))

# Calculate the range (max-min) and print it
print('range: ' + str(df['Petal Length'].ptp()))


print('\n\nPetal Width:\n')
print(df['Petal Width'].describe())

# Calculate the variance (square of std) and print it
print('variance: ' + str(df['Petal Width'].var()))

# Calculate the range (max-min) and print it
print('range: ' + str(df['Petal Width'].ptp()))
```

#### 1.2 Data Visualization

##### Histograms

```python
# Adapted from Paul H. (https://stackoverflow.com/questions/25539195/multiple-histograms-in-pandas)

fig, ax = plt.subplots()
ax.set_title("Sepal Width and Length Histogram")

a_heights, a_bins = np.histogram(df['Sepal Width'])
b_heights, b_bins = np.histogram(df['Sepal Length'])

width = (a_bins[1] - a_bins[0])/3

ax.bar(a_bins[:-1], a_heights, width=width, facecolor='cornflowerblue', label='Width')
ax.bar(b_bins[:-1]+width, b_heights, width=width, facecolor='seagreen', label='Length')
ax.legend()



fig, ax = plt.subplots()
ax.set_title("Petal Width and Length Histogram")

a_heights, a_bins = np.histogram(df['Petal Width'])
b_heights, b_bins = np.histogram(df['Petal Length'])

width = (a_bins[1] - a_bins[0])/3

ax.bar(a_bins[:-1], a_heights, width=width, facecolor='darkgoldenrod', label='Width')
ax.bar(b_bins[:-1] + width, b_heights, width=width, facecolor='mediumpurple', label='Length')
ax.legend()
```

##### Boxplot

```python
boxprops = dict(linestyle='-', linewidth=2, color='k')
medianprops = dict(linestyle='-', linewidth=2, color='k')
box = df.boxplot(grid=False, boxprops=boxprops, medianprops=medianprops, figsize=(6,5), return_type='axes')
box.set_title('Boxplot of Iris Dataset')
```

#### 2 Pen Digits Dataset

```python
fileURL = "pendigits/pendigits.tra"
pendigit = pd.read_csv(fileURL, names=["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10", "P11", "P12", "P13", "P14", "P15", "P16", "Digit"], header=None)
```

#### 2.1 Summary Statistics

```python
# Print the first 5 elements
pendigit.head(5)
```

##### There are sixteen features and one class.
##### The sixteen features are numeric (interval) from 0-100 measuring accuracy to T points (n=8) of the digit being drawn. T is resampled for a total of 16 measures.
##### The last column is categorical (nominal) and describes the class of digit (0-9) being drawn.

```python
# Wanted to use bold inside my code output to make things a bit more readable. Found this snippet by Jakob (https://stackoverflow.com/questions/32026727/format-output-of-code-cell-with-markdown)
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))

# Since we need to do this print feature stats 16 times, write a function once and use that
def printPoint(number):
    printmd('**'+ 'P' + str(number) +':**')
    print(pendigit['P' + str(number)].describe())

    # Calculate the variance (square of std) and print it
    print('variance: ' + str(pendigit['P' + str(number)].var()))

    # Calculate the range (max-min) and print it
    print('range: ' + str(pendigit['P' + str(number)].ptp() + "\n"))
    

for x in range(1, 17):
    printPoint(x)
```

#### 2.2 Data Visualization

##### Histograms

```python

colors = ['forestgreen', 'indigo', 'lavender', 'lightslategrey',
          'mediumorchid', 'orangered', 'powderblue', 'tomato',
          'yellow', 'turquoise', 'gold', 'darkkhaki',
          'crimson', 'lime', 'midnightblue', 'mediumvioletred']

def histoPoint(number):
    fig, ax = plt.subplots()
    ax.set_title('Point ' + str(number) + ' histogram')

    a_heights, a_bins = np.histogram(pendigit['P' + str(number)])
    width = (a_bins[1] - a_bins[0])/3

    ax.bar(a_bins[:-1], a_heights, width=width, facecolor=colors[number-1], label='Point ' + str(number))
    ax.legend()
    
for x in range(1, 17):
    histoPoint(x)
    
```

##### Boxplot

```python
boxprops = dict(linestyle='-', linewidth=2, color='k')
medianprops = dict(linestyle='-', linewidth=2, color='k')
box = pendigit.boxplot(grid=False, boxprops=boxprops, medianprops=medianprops, figsize=(10,5), return_type='axes')
box.set_title('Boxplot of Pen Digit Dataset')
```
